The data is located in `../datasets/penguins_classification.csv`, load it with

```python
# Write your code here.
import pandas as pd
df=pd.read_csv('../data/penguins_classification.csv')
df.head()
```

Show a few samples of the data.

How many features are numerical? How many features are categorical?

```python
# Write your code here.
num_cols = df.select_dtypes(include=['number']).shape[1]
cat_cols = df.select_dtypes(include=['object', "category"]).shape[1]
print(f'The number of numerical columns is: {num_cols}')
print(f'The number of categorical columns is: {cat_cols}')
```
